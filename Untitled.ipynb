```python
#I will use pandas to do data transforming, manipulation, and analysis

import pandas as pd
```

```python
#I chose a topic which is quite interesting for me, since I have been binging k-dramas recently
#I want to know some things from the kdramas, and this csv I found at Kaggle.com seemed to be a good choice

df = pd.read_csv('top100_kdrama.csv')
```

```python
df.head()

#You can see what kind of attributes we have here
#I believe I will need to clean the data, since I can't really see how to get use out of the Synopsis column
#The Start_date is not in the best format neither, I also think I will only need the year from that - not quite sure yet
#The Tags seems to be not that useful either, I might have to remove that too
#I can see some NaN data in End_date, so those are the ongoing dramas
#I might need to modify the Genre column too, or at least some of the data, because each drama has more han 1 genres
#Maybe I could also seperate the Main Role column into two columns - Female and Male roles 
#this might be problematic beacause koreans use certain names for women and men also. 
#Even,  I would need to use classification to determine these - It would be another fun project to do :) 
```

```python
#Let's remove the Synopsis column
#We need axis=1 so that pandas knows it is about columns

df.drop('Synopsis', axis=1, inplace=True)
df.head()
```

```python
#I will also need to extract the year from Start_date
#An easy way to do so is by using DateTimeIndex feature

df['Start_year'] = pd.DatetimeIndex(df['Start_date']).year.astype(str)
df.head()
```

```python
#I don't see how I will use the Tags data, so I might drop that column too

df.drop('Tags', axis=1, inplace=True)
df.head()
```

```python
#I wanted to modify Genre too, but I think I will leave it as it is for now
#The Watcher is in bad format, so I need modify that - no need to create a new column for that
#the lambda expression is perfect use for this - applying only to the Watchers column
#And also need to convert that type to int

df.Watchers = df.Watchers.apply(lambda x : x.replace(',','')).astype(int)
```

```python
#Let's start the analysis :)
df.head()
```

```python
#Is the number of episodes in relation of popularity or rank or the number of watchers?
#I need to create a subcollection of the attributes I want to analyze

sub_of_duration_and_rank = df[["Duration","Rank","Popularity", "Watchers", "Episodes"]]
```

```python
#then I check how these datas correlate with each other

sub_of_duration_and_rank.corr()
```

```python
# We can see that :
# Duration is in weak negatice corr with Rank
# Duration is in weak negative corr with Popularity
# Duration is in weak positive corr with Watchers
# Duration is in weak negative corr with Episodes
# Rank is in moderate positive corr with Popularity
# Rank is in weak negative corr with Watchers
# Rank is in wek positive corr with Episodes
# Popularity is in strong negative corr with Watchers
# Meaning, that the Popularity and the number of Watchers are moving strongly in the opposite direction in a linear
# So by having high number of watchers, It does not mean the show has a high popularity rank
# Also, the Rank and Popularity is in moderate positive correlation, so the higher the Rank, higher the Popularity
```

```python
#Are the number of watchers grow by year ? = Are K-Dramas more popular recently?
#For these I only need the Watchers and the Start_year 
#I will sum up the number of watchers for each year and see how it's changing

watchers_by_year = df.groupby(["Start_year"])["Watchers"].sum()
watchers_by_year

#I checked if the numbers are good - they are
#filtered_df = df.loc[df['Start_year'] == 2015]
```

```python
#I would like to expand this with the number of dramas released that year

number_of_dramas_per_year = df.groupby(["Start_year"]).nunique()['ID']

#Now I need to merge these together

result = pd.concat([number_of_dramas_per_year, watchers_by_year], axis=1, join='inner')
result
```

```python
#For me, the ID is quite unlogical column name here, let's try to rename that

result = result.rename(columns={'ID' : 'Dramas_released'})
result
```

```python
#Find out which year was the most productive for drama makers

max_dramas = result.loc[result['Dramas_released'].idxmax()]
max_dramas
```

```python
#In which year is the watchers' num is the highest?

max_watchers = result.loc[result['Watchers'].idxmax()]
max_watchers
```

```python
#So in 2022 were the most dramas released but it was not the year when the watchers' number was the highest - it was 2021
```

```python
#Maybe we could also check which actor was the most popular - how many dramas are they in
#I could save the names of actors in a list, and then check the occurences from that, since the main role has more than 1 name


names = df['Main Role'].str.split(', ').tolist()

#Now i have a really ugly list in a list in a list..
```

```python
#I want them to be one single list, and also remove the duplicates 

unique_names = list(set().union(*names))

len(unique_names)
```

```python
#Now I have a long list (299 names) of the actors - see who had the most part in dramas throghout the years

actor_counts = {name: 0 for name in unique_names}

# Count the occurrences of each actor in the "Main Role" column
for names in df['Main Role']:
    for name in unique_names:
        if name in names:
            actor_counts[name] += 1
            
#let's make a df from it - much more readable than storing it in a list
actors_df = pd.DataFrame(list(actor_counts.items()), columns=['Actor', 'Count'])


# Display the actor counts
print(actors_df)
```

```python
#Which actor was most seen in dramas?

actors_df.sort_values(by=['Count'])
```

```python
#The actor was in 6 dramas, and he is Lee Joon

lee_dramas = df.where(df["Main Role"].str.contains("Lee Joon")).dropna()

#Which of his drama has the highest score ?

lee_dramas.where(lee_dramas["Score"] == lee_dramas["Score"].max()).dropna()
```
